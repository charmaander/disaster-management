```python
%%writefile weather_app.py
import streamlit as st
import requests
import numpy as np

# --- Weather Data Fetching Function ---
def fetch_live_weather_data(location):
    api_key = '64f8daf9baac8e5424bf7870eda86287'  # Replace with your Weatherstack API key

    # Construct the URL for the Weatherstack API request
    url = f'http://api.weatherstack.com/current?access_key={api_key}&query={location}'

    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        temp = data['current']['temperature']
        humidity = data['current']['humidity']
        rainfall = data['current'].get('precip', 0)

        return temp, humidity, rainfall

    except requests.exceptions.HTTPError as err:
        st.error(f"HTTP error occurred: {err}")
        return None, None, None
    except Exception as err:
        st.error(f"An error occurred: {err}")
        return None, None, None

# --- Streamlit App ---
st.title("Live Weather Dashboard")

location = st.text_input("Enter a location (e.g., New York, London)", "New York")

if st.button("Get Weather"):
    temp, humidity, rainfall = fetch_live_weather_data(location)

    if temp is not None:
        st.write(f"**Temperature:** {temp}°C")
        st.write(f"**Humidity:** {humidity}%")
        st.write(f"**Rainfall:** {rainfall} mm")
    else:
        st.write("Could not retrieve weather data for the specified location.")
```

```python
import requests

def get_location_key(city_name, api_key):
    # URL for searching location by city name
    url = f"http://dataservice.accuweather.com/locations/v1/cities/search?apikey={api_key}&q={city_name}"

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if data:
            location_key = data[0]['Key']  # Get the Location Key of the first search result
            print(f"Location Key for {city_name}: {location_key}")
            return location_key
        else:
            print("Location not found.")
            return None
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None
```

```python
app = Flask(__name__)

# Define a function to fetch live data from Google Earth Engine
def fetch_live_data(latitude, longitude):
    try:
        # Define time range for live data
        end_date = datetime.date.today()
        start_date = end_date - datetime.timedelta(days=1)  # Last 24 hours

        # Define the region of interest
        region = ee.Geometry.Point([longitude, latitude])

        # Fetch data from Google Earth Engine (using CHIRPS dataset for precipitation as example)
        dataset = ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY") \
            .filterDate(str(start_date), str(end_date)) \
            .filterBounds(region)

        # Calculate mean precipitation over the past day
        precipitation = dataset.mean().reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=region,
            scale=1000
        )

        # Fetch the result
        precip_data = precipitation.getInfo().get('precipitation', 0)

        # Repeat for temperature (using ERA5 data as an example)
        temperature_dataset = ee.ImageCollection("ECMWF/ERA5/DAILY") \
            .filterDate(str(start_date), str(end_date)) \
            .filterBounds(region)

        temperature = temperature_dataset.mean().reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=region,
            scale=1000
        )

        temp_data = temperature.getInfo().get('mean_2m_air_temperature', 0)

        # Organize data into a list for model prediction input
        return [temp_data, precip_data]
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None

# Define endpoint for prediction
@app.route('/predict', methods=['POST'])
def predict():
    # Get input data from request
    data = request.get_json()
    latitude = data.get('latitude')
    longitude = data.get('longitude')

    # Fetch live data
    live_data = fetch_live_data(latitude, longitude)
    if live_data is None:
        return jsonify({'error': 'Unable to fetch live data'}), 500

    # Format data as required by the model
    input_data = np.array([live_data])  # Reshape to match model input shape

    # Perform prediction
    prediction = model.predict(input_data)

    # Return prediction result
    return jsonify({'prediction': prediction.tolist()})

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
```
